```python
def add_demographic(data_input):
    if data_input.Age >=20:
        demo = "adut_m" if data_input.Sex == "male" else "adult_f"
    else:
        demo = "boy_and_girl"

    return demo

data.ix[ : , "Demographic"] = data.apply(add_demographic, axis=1)
```
